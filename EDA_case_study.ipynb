```python
# i want to create an income band from estimated income column
bins = [0, 100000, 300000,float('inf')]
labels=['Low','Medium','High']  
df['Income Band'] = pd.cut(df['Estimated Income'], bins=bins, labels=labels)
```

```python
df['Income Band'].value_counts().plot(kind='bar')
```

```python
df.columns
```

```python
#examine the distribution of unique values  cataegories in categorical columns
categorical_cols = df[[ 'BRId', 'GenderId', 'IAId', 'Amount of Credit Cards','Nationality', 'Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting', 'Income Band',
                       ]]

for col in categorical_cols:
    print(f"Value counts for column {col}:")
    print(df[col].value_counts())
    print("\n")
    display(df[col].value_counts().plot(kind='bar'))
    plt.show()
```

Univariate Analysis

```python
for i ,predictor in enumerate(df[[ 'BRId', 'GenderId', 'IAId', 'Amount of Credit Cards','Nationality', 'Occupation','Fee Structure','Loyalty Classification','Properties Owned','Risk Weighting', 'Income Band',
                       ]]):
    plt.figure(i)
    sns.countplot(x=predictor, data=df)
```

```python
#Histplot of value Counts for different occupations

for col in categorical_cols:
    if col == 'Occupation':
        continue
    plt.figure(figsize=(8,4))
    sns.histplot(df[col])
    plt.title('Histogram of Occupation Count')
    plt.xlabel(col)
    plt.ylabel('Count')
    plt.show()
    
    
```

Numerical analysis

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans',
       'Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending']
for col in numerical_cols:
    plt.figure(figsize=(8,4))
    sns.histplot(df[col], kde=True)
    plt.title(f'Histogram of {col}')
    plt.xlabel(col)
    plt.ylabel('Frequency')
    plt.show()
```

Heatmaps

```python
numerical_cols = ['Estimated Income', 'Superannuation Savings','Credit Card Balance','Bank Loans',
       'Bank Deposits', 'Checking Accounts', 'Saving Accounts','Foreign Currency Account', 'Business Lending']

correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Insight of EDA

1.The strongest positive correlation occur among 'Bank Deposits with Checking Accounts', and "Forigen Currency Account " indicating that customers who maintain high balances in one account type often hold subtantial amount / funds across other accounts as well.
